### Problema





> <div style="text-align: justify"> A raíz de la pandemia ocasionada por el Covid-19 las empresas, específicamente las MIPyMES, se han enfrentado a muchas dificultades mismas que han sido fuentes de estrés, la mayoría relacionada con el ámbito económico. Por esta razón, se consideró al estrés como un concepto relevante en el presente trabajo.</div>


> <div style="text-align: justify"> Se denominó estresor económico a todo estímulo y situaciones relacionadas con el medio laboral y económico que puede llegar a recibir una empresa, generando una inestabilidad económica en la misma, tales como despidos, pago de salarios, disminución en las ventas, etc. Estos factores externos en conjunto provocan en la empresa lo que llamamos: estrés económico.</div>


> <div style="text-align: justify"> Las empresas pueden tener estrés económico en mayor o menor medida, el problema es que esto resulta ser algo intangible por consiguiente algo muy difícil de ver. Con esta métrica se espera que los usuarios sean capaces de reconocer cuando un negocio está pasando por mucho estrés económico y así poder identificar mejor cuáles son las problemáticas a atacar y los pasos a seguir para que esta empresa pueda encontrar estabilidad.</div>

### Criterio de Construcción de Variable

> <div style="text-align: justify"> Para construir esta variable lo primero que se realizó fue una escala para poder evaluar los estresores económicos en ella. Se consideró pertinente dividir la frecuencia del estrés en una escala de cinco puntos del 0 al 4 basada en una prueba psicométrica, donde 0 representa una nula o menor cantidad de estrés y el cuatro se considera como gran cantidad de estrés. Además se decidió agregar dos nuevos estados -1 y -2 que representan no solo que la empresa no tiene estrés sino que demuestra tener tranquilidad y por consecuente una estabilidad en el ámbito económico.</div>


| <div style="text-align: center">-2                                            |<div style="text-align: center"> -1                                  | <div style="text-align: center">0                                           |<div style="text-align: center"> 1                                               | <div style="text-align: center">2                                                   | <div style="text-align: center">3                                                       | <div style="text-align: center">4                                    |
|-----------------------------------------------|-------------------------------------|---------------------------------------------|-------------------------------------------------|-----------------------------------------------------|---------------------------------------------------------|--------------------------------------|
|<div style="text-align: left"> Se manifiesta tener mucha e<br><br>stabilidad</div> |<div style="text-align: left"> Se manifiesta<br> tener estabilidad </div>|<div style="text-align: left"> No se manifiesta alguna <br> respuesta de estrés</div> |<div style="text-align: left"> Raramente se considera <br> como situación de estrés</div> |<div style="text-align: left"> En ocasiones/a veces se <br>considera estar bajo estrés</div> |<div style="text-align: left"> La respuesta de <br>estrés se considera <br> como algo frecuente</div> |<div style="text-align: left"> Se considera estar <br> bajo mucho estrés </div>|
    
> <div style="text-align: justify"> El siguiente paso fue clasificar las variables contenidas en la base de datos utilizada como estresores económicos cuando fuera el caso. </div>

> <div style="text-align: justify">A continuación se muestran las variables que se obtuvieron de la base de datos: “Diagnóstico de las Afectaciones Económicas a las Unidades Económicas de Jalisco por el Coronavirus” misma que se extrajo de encuestas enviadas por correo electrónico y contiene valiosa información de algunas empresas de Jalisco. </div>


> <div style="text-align: justify">Las siguientes variables se consideraron como estresores económicos ya que engloban  situaciones relacionadas con el medio laboral y económico, en donde, en este caso, las empresas se vieron en la necesidad de afrontar ciertas pérdidas o cambios con respecto a su actividad económica. Tales como los despidos o pagos de salarios, situaciones que pueden llegar a causar estrés tomando en cuenta los estímulos y factores externos, en este caso se considera como factor externo la situación de la pandemia.  </div>


> <div style="text-align: justify">Las variables proporcionadas y clasificadas como estresores económicos nos brindan información y datos cualitativos, por lo que se propuso una escala, mencionada anteriormente, que se aplicó a cada variable y su naturaleza para que puedan ser cuantificables y así poder estimar la variable estrés por cada empresa que participó en esta encuesta. </div>


> <div style="text-align: justify">La variable estrés al final será el valor resultante de la suma  de las variables seleccionadas y su escala que se detalla en el siguiente recuadro. Es importante recalcar que cuanto más alto sea ese valor más alto será el estrés de la empresa. </div>


> <div style="text-align: justify">Nota: Las variables relacionadas con pagos fijos cuyas respuestas están dada en semanas, para el cálculo de la métrica se tomaron como una sola. </div>

### Construcción de variable

```python
#Importar Librerías 
import pandas as pd 
import numpy as np
```

### Base de datos: IIEG_E_1

```python
#Lectura de la base de datos
data=pd.read_excel('data.xlsx', sheet_name='IIEG_E_1') 
#Acomodo de datos
data = data.replace({"Más de 52": 52, "De 26 a 52": 26, "No aplica": 100, "No contesto": 101, "No sé":102, "Más de un año":12})
#Variable a predecir 
estres = np.zeros(len(data))
# Variable global de pagos
pagos = np.zeros(len(data))
```

```python
#Calculos descritos en la tabla anterior

#ventas_porcentaje
for i in range(0, len(data)):
    if data['ventas_porcentaje'][i]>0 and data['ventas_porcentaje'][i]<=25:
        estres[i] += 1
    if data['ventas_porcentaje'][i]>25 and data['ventas_porcentaje'][i]<=50:
        estres[i] += 2
    if data['ventas_porcentaje'][i]>50 and data['ventas_porcentaje'][i]<=75:
        estres[i] += 3
    if data['ventas_porcentaje'][i]>75 and data['ventas_porcentaje'][i]<=100:
        estres[i] += 4
        
#perdidas_porcentaje
for i in range(0, len(data)):
    if data['perdidas_porcentaje'][i]>0 and data['perdidas_porcentaje'][i]<=25:
        estres[i] += 1
    if data['perdidas_porcentaje'][i]>25 and data['perdidas_porcentaje'][i]<=50:
        estres[i] += 2
    if data['perdidas_porcentaje'][i]>50 and data['perdidas_porcentaje'][i]<=75:
        estres[i] += 3
    if data['perdidas_porcentaje'][i]>75 and data['perdidas_porcentaje'][i]<=100:
        estres[i] += 4
        
#capacidad
for i in range(0, len(data)):
    if data['capacidad'][i]>0 and data['capacidad'][i]<=25:
        estres[i] += 1
    if data['capacidad'][i]>25 and data['capacidad'][i]<=50:
        estres[i] += 2
    if data['capacidad'][i]>50 and data['capacidad'][i]<=75:
        estres[i] += 3
    if data['capacidad'][i]>75 and data['capacidad'][i]<=100:
        estres[i] += 4
        
#cerrar
for i in range(0, len(data)):
    if data['cerrar'][i] == 'Sí':
        estres[i] += 2
    if data['cerrar'][i] == 'No':
        estres[i] += 0
        
#despidos
for i in range(0, len(data)):
    if data['despidos'][i] == 'Sí' or data['despidos'][i] == 'No, pero lo está considerando':
        estres[i] += 2
    if data['despidos'][i] == 'No' or data['despidos'][i] == 'No cuenta con personal':
        estres[i] += 0
    if data['despidos'][i] == 'No, pero lo hará en los próximos días':
        estres[i] += 3
        
#plantilla_porcentaje
for i in range(0, len(data)):
    if data['plantilla_porcentaje'][i]>0 and data['plantilla_porcentaje'][i]<=25:
        estres[i] += 1
    if data['plantilla_porcentaje'][i]>25 and data['plantilla_porcentaje'][i]<=50:
        estres[i] += 2
    if data['plantilla_porcentaje'][i]>50 and data['plantilla_porcentaje'][i]<=75:
        estres[i] += 3
    if data['plantilla_porcentaje'][i]>75 and data['plantilla_porcentaje'][i]<=100:
        estres[i] += 4
        
#capacidad_pago
for i in range(0, len(data)):
    if data['capacidad_pago'][i] == 'Sí':
        estres[i] += 1
    if data['capacidad_pago'][i] == 'No':
        estres[i] += 2
    if data['capacidad_pago'][i] == 'No sé':
        estres[i] += 0

#credito
for i in range(0, len(data)):
    if data['credito'][i] == 'Sí':
        estres[i] += 2
    if data['credito'][i] == 'No':
        estres[i] += 0
        
#aumento_insumos
for i in range(0, len(data)):
    if data['aumento_insumos'][i] == 'Sí':
        estres[i] += 3
    if data['aumento_insumos'][i] == 'No':
        estres[i] += 0
        
#aumento_precios
for i in range(0, len(data)):
    if data['aumento_precios'][i] == 'Sí':
        estres[i] += 3
    if data['aumento_precios'][i] == 'No':
        estres[i] += 0
        
#escasez_insumos
for i in range(0, len(data)):
    if data['escasez_insumos'][i] == 'Sí':
        estres[i] += 3
    if data['escasez_insumos'][i] == 'No':
        estres[i] += 0
        
#aumento_tipocambio
for i in range(0, len(data)):
    if data['aumento_tipocambio'][i] == 'Sí':
        estres[i] += 4
    if data['aumento_tipocambio'][i] == 'No':
        estres[i] += 0
```

```python
#Calculo de la variable global de pagos

#pago_salarios
pagos_salario = np.zeros(len(data))
for i in range(0, len(data)):
    if data['pago_salarios'][i] == 100 or data['pago_salarios'][i] == 101 or data['pago_salarios'][i] == 102:
        pagos_salario[i] += 0
    if data['pago_salarios'][i]>0 and data['pago_salarios'][i]<=6:
        pagos_salario[i] += 4
    if data['pago_salarios'][i]>6 and data['pago_salarios'][i]<=12:
        pagos_salario[i] += 3
    if data['pago_salarios'][i]>12 and data['pago_salarios'][i]<=18:
        pagos_salario[i] += 2
    if data['pago_salarios'][i]>18:
        pagos_salario[i] += -1
        
#pago_impuestos
pagos_impuestos = np.zeros(len(data))
for i in range(0, len(data)):
    if data['pago_impuestos'][i] == 100 or data['pago_impuestos'][i] == 101 or data['pago_impuestos'][i] == 102:
        pagos_impuestos[i] += 0
    if data['pago_impuestos'][i]>0 and data['pago_impuestos'][i]<=6:
        pagos_impuestos[i] += 4
    if data['pago_impuestos'][i]>6 and data['pago_impuestos'][i]<=12:
        pagos_impuestos[i] += 3
    if data['pago_impuestos'][i]>12 and data['pago_impuestos'][i]<=18:
        pagos_impuestos[i] += 2
    if data['pago_impuestos'][i]>18:
        pagos_impuestos[i] += -1
        
#pago_credito
pagos_credito = np.zeros(len(data))
for i in range(0, len(data)):
    if data['pago_creditos'][i] == 100 or data['pago_creditos'][i] == 101  or data['pago_creditos'][i] == 102:
        pagos_credito[i] += 0
    if data['pago_creditos'][i]>0 and data['pago_creditos'][i]<=6:
        pagos_credito[i] += 4
    if data['pago_creditos'][i]>6 and data['pago_creditos'][i]<=12:
        pagos_credito[i] += 3
    if data['pago_creditos'][i]>12 and data['pago_creditos'][i]<=18:
        pagos_credito[i] += 2
    if data['pago_creditos'][i]>18:
        pagos_credito[i] += -2
        
#pago_renta
pagos_renta = np.zeros(len(data))
for i in range(0, len(data)):
    if data['pago_renta'][i] == 100 or data['pago_renta'][i] == 101  or data['pago_renta'][i] == 102:
        pagos_renta[i] += 0
    if data['pago_renta'][i]>0 and data['pago_renta'][i]<=6:
        pagos_renta[i] += 4
    if data['pago_renta'][i]>6 and data['pago_renta'][i]<=12:
        pagos_renta[i] += 3
    if data['pago_renta'][i]>12 and data['pago_renta'][i]<=18:
        pagos_renta[i] += 2
    if data['pago_renta'][i]>18:
        pagos_renta[i] += -2
        
#pago_internet
pagos_internet = np.zeros(len(data))
for i in range(0, len(data)):
    if data['pago_internet'][i] == 100 or data['pago_internet'][i] == 101  or data['pago_internet'][i] == 102:
        pagos_internet[i] += 0
    if data['pago_internet'][i]>0 and data['pago_internet'][i]<=6:
        pagos_internet[i] += 4
    if data['pago_internet'][i]>6 and data['pago_internet'][i]<=12:
        pagos_internet[i] += 3
    if data['pago_internet'][i]>12 and data['pago_internet'][i]<=18:
        pagos_internet[i] += 2
    if data['pago_internet'][i]>18:
        pagos_internet[i] += -2
        
#pago_agua
pagos_agua = np.zeros(len(data))
for i in range(0, len(data)):
    if data['pago_agua'][i] == 100 or data['pago_agua'][i] == 101  or data['pago_agua'][i] == 102:
        pagos_agua[i] += 0
    if data['pago_agua'][i]>0 and data['pago_agua'][i]<=6:
        pagos_agua[i] += 4
    if data['pago_agua'][i]>6 and data['pago_agua'][i]<=12:
        pagos_agua[i] += 3
    if data['pago_agua'][i]>12 and data['pago_agua'][i]<=18:
        pagos_agua[i] += 2
    if data['pago_agua'][i]>18:
        pagos_agua[i] += -2

#pago_luz
pagos_luz = np.zeros(len(data))
for i in range(0, len(data)):
    if data['pago_luz'][i] == 100 or data['pago_luz'][i] == 101  or data['pago_luz'][i] == 102:
        pagos_luz[i] += 0
    if data['pago_luz'][i]>0 and data['pago_luz'][i]<=6:
        pagos_luz[i] += 4
    if data['pago_luz'][i]>6 and data['pago_luz'][i]<=12:
        pagos_luz[i] += 3
    if data['pago_luz'][i]>12 and data['pago_luz'][i]<=18:
        pagos_luz[i] += 2
    if data['pago_luz'][i]>18:
        pagos_luz[i] += -2

#pago_gas
pagos_gas = np.zeros(len(data))
for i in range(0, len(data)):
    if data['pago_gas'][i] == 100 or data['pago_gas'][i] == 101  or data['pago_gas'][i] == 102:
        pagos_gas[i] += 0
    if data['pago_gas'][i]>0 and data['pago_gas'][i]<=6:
        pagos_gas[i] += 4
    if data['pago_gas'][i]>6 and data['pago_gas'][i]<=12:
        pagos_gas[i] += 3
    if data['pago_gas'][i]>12 and data['pago_gas'][i]<=18:
        pagos_gas[i] += 2
    if data['pago_gas'][i]>18:
        pagos_gas[i] += -2
        
#duracion
duracion = np.zeros(len(data))
for i in range(0, len(data)):
    if  data['duracion'][i] == 12 or data['duracion'][i] == 102:
        duracion[i] += 0
    if data['duracion'][i]>0 and data['duracion'][i]<=6:
        duracion[i] += 4
    if data['duracion'][i]>6 and data['duracion'][i]<=12:
        duracion[i] += 3
    if data['duracion'][i]>12 and data['duracion'][i]<=18:
        duracion[i] += 2
    if data['duracion'][i]>18:
        duracion[i] += -2
        
#Definir el valor de cada variable que compone pagos
p_especificos = .15
p_general = .1
#Calculo de estres en la variable pagos
for i in range(0, len(data)):
    pagos[i] = pagos_salario[i]*p_especificos + duracion[i]*p_especificos + pagos_impuestos[i] * p_general + \
    pagos_credito[i] * p_general + pagos_renta[i] * p_general + pagos_internet[i] * p_general + pagos_agua[i] * p_general + \
    pagos_luz[i] * p_general +  pagos_gas[i] * p_general
```

### Resultados

> El resultado es la misma base de datos con la nueva columna que contiene el valor del estrés para cada empresa.

```python
# Agregar pagos a la variable estres
estres = estres + pagos

# Agregar la variable al DataFrame
data['Estrés'] = estres

#Visualización de la variable
data
```
